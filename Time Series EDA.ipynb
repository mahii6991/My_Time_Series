```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla= pdr.get_data_yahoo('TSLA')#getting the stock price of tesla from today to past 5 years
```

```python
df_tesla.head()#viewing the head of the dataset,as it starts from 2017
```

```python
df_tesla.tail()#we can see we have the dastaset till now
```

```python
df_tesla.plot(figsize=(12,4))#plotting the dataset and viewing the volatility of the dataset
```

```python
df_tesla['High'].plot(figsize=(12,4))#plotting only the highs of the dataset
```

```python
#now we will going to zoom some datapoints in the graph
#so we will going to set some x lim and y lim in the dataset
df_tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],figsize=(12,4))
#in this way we have magnified the datapoints that we want to see
```

```python
#setting the ylim as well
df_tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],ylim=[0,900],figsize=(12,4))
```

```python
#changing the color of the line
df_tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],ylim=[0,900],figsize=(12,4),c="green")
```
